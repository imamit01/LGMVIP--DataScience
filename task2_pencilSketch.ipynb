### Imprt Module

```python
import cv2
import matplotlib.pyplot as plt
```

### Read Image file

```python
image=cv2.imread("img.jpg")
```

### Plot the Image

```python
plt.imshow(image[:,:,::-1])
plt.axis(False)
plt.show()
```

### Read the image in RBG format

```python
RGB_Image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(RGB_Image)
plt.axis(False)
plt.show()
```

### Convert it to a grayscale image

```python
gray_image= cv2.cvtColor(RGB_Image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_image , cmap="gray")
plt.axis(False)
plt.show()
```

### Inverted Image

```python
inverted = 255-gray_image
plt.imshow(inverted , cmap="gray")
plt.axis(False)
plt.show()
```

### Smoothen Image

```python
Smoothen_img = cv2.GaussianBlur(inverted, (21, 21), 0)
plt.imshow(Smoothen_img , cmap="gray")
plt.axis(False)
plt.show()
```

### Converting images in pencil sketches

```python
Pencil_Sketch = cv2.divide(gray_image, 255 - Smoothen_img, scale=255)
plt.imshow(Pencil_Sketch,cmap="gray")
plt.axis(False)
plt.show()
```
